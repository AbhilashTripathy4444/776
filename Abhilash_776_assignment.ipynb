```python
import pandas as pd
population_india_census2011=pd.read_excel("population_india_census2011.xlsx")
population_india_census2011
```

```python
import numpy as np
df_morph = population_india_census2011[[ "Population", "Gender Ratio", "Rural population","Urban population"]].drop_duplicates()
df_morph.shape
```

```python
boxplot = df_morph.boxplot([ 'Population','Gender Ratio','Rural population','Urban population'])
```

```python
import matplotlib.pyplot as plt

population_india_census2011['Population'].plot(kind='pie',
                            figsize=(19,8),
                            autopct='%1.1f%%', 
                            startangle=60,  
                            pctdistance=1.12,
                            labels=None,   
                            )

plt.title('Total Population according to State')
plt.legend(labels=population_india_census2011.index, loc='upper left') 
plt.axis('equal')

plt.show()
```
